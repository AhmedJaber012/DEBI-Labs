# SQL Queries Overview with Python and SQLite3

In this notebook, we'll explore different types of SQL queries using Python's `sqlite3` module. We'll cover topics ranging from basic queries to more advanced concepts like joins, unions, and views. Each section will include explanations and code examples.


## Setting Up SQLite3 in Python

```python
import sqlite3

# Create an in-memory database for demonstration purposes
connection = sqlite3.connect(':memory:')
cursor = connection.cursor()

# Create sample tables for the examples
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT NOT NULL,
    student_id INTEGER,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
)
''')

# Insert sample data into the students table
cursor.executemany('''
INSERT INTO students (name, age) VALUES (?, ?)
''', [('Alice', 22), ('Bob', 24), ('Charlie', 20)])

# Insert sample data into the courses table
cursor.executemany('''
INSERT INTO courses (course_name, student_id) VALUES (?, ?)
''', [('Mathematics', 1), ('Physics', 2), ('Biology', 1), ('Chemistry', 3)])


```

## Basic SQL Queries

```python
# SELECT query to fetch all data from the students table
cursor.execute('SELECT * FROM students')
students = cursor.fetchall()

print("Students Table:")
for student in students:
    print(student)

# # SELECT query to fetch specific columns
cursor.execute('SELECT name, age FROM students')
names_and_ages = cursor.fetchall()

print("\nNames and Ages:")
for entry in names_and_ages:
    print(entry)
```

## Filtering Data with WHERE Clause


```python
# SELECT with WHERE clause
cursor.execute('SELECT * FROM students WHERE age > 21')
filtered_students = cursor.fetchall()

print("Students older than 21:")
for student in filtered_students:
    print(student)
```

## Aggregate Functions (COUNT, AVG, SUM)

```python
# COUNT query
cursor.execute('SELECT COUNT(*) FROM students')
count = cursor.fetchone()[0]
print(f"Total number of students: {count}")

# AVG query
cursor.execute('SELECT AVG(age) FROM students')
average_age = cursor.fetchone()[0]
print(f"Average age of students: {average_age:.2f}")
```

## GROUP BY and HAVING Clauses
- **GROUP BY** is used to aggregate data across rows with the same values in specified columns.
- **HAVING** filters the results of `GROUP BY`, similar to `WHERE` but applied to grouped rows.


```python
# GROUP BY query to count courses per student
cursor.execute('SELECT student_id, COUNT(course_id) FROM courses GROUP BY student_id')
course_counts = cursor.fetchall()

print("Courses taken by each student:")
for student_id, count in course_counts:
    print(f"Student ID {student_id} has taken {count} course(s).")
```

```python
# GROUP BY query to count courses per student and filter with HAVING
cursor.execute('''
SELECT student_id, COUNT(course_id) as course_count
FROM courses
GROUP BY student_id
HAVING course_count > 1
''')
having_example = cursor.fetchall()

print("Students with more than one course (GROUP BY with HAVING):")
for student_id, count in having_example:
    print(f"Student ID {student_id} has taken {count} course(s).")
```

## Sorting with ORDER BY
- **ORDER BY** sorts the results of a query based on one or more columns.
- You can sort in **ascending order** (default) or **descending order** using `ASC` or `DESC`.


```python
# SELECT query with ORDER BY using multiple columns
cursor.execute('''
SELECT * FROM students
ORDER BY age DESC, name ASC
''')
sorted_students_multi = cursor.fetchall()

print("Students sorted by age (descending) and then by name (ascending):")
for student in sorted_students_multi:
    print(student)
```

## JOIN Operations

### Intuition:
In SQL, a **JOIN** is used to combine rows from two or more tables based on a related column. Think of it as merging information from different sources to answer questions like "Which students are taking which courses?"

### Types of Joins:
- **INNER JOIN**: Only returns rows where there is a match in both tables.
- **LEFT JOIN (LEFT OUTER JOIN)**: Returns all rows from the left table and matched rows from the right table. If no match is found, NULLs are returned for columns from the right table.
- **RIGHT JOIN**: Like LEFT JOIN but includes all rows from the right table. (Note: SQLite doesn't support RIGHT JOIN directly, but the effect can be achieved by switching table order in LEFT JOIN.)
- **FULL OUTER JOIN**: Returns rows when there is a match in one of the tables. (Not supported directly by SQLite; workarounds involve combining LEFT and RIGHT JOIN results.)

Now, let's look at some JOIN examples:


```python
# INNER JOIN to combine students and courses
cursor.execute('''
SELECT students.student_id, students.name, courses.course_name
FROM students
INNER JOIN courses ON students.student_id = courses.student_id
''')
joined_data = cursor.fetchall()
print(joined_data)
print("Students and their courses:")
for name, course in joined_data:
    print(f"{name} is enrolled in {course}")
```

```python
# INNER JOIN to combine students and courses where there is a match
cursor.execute('''
SELECT students.name, courses.course_name
FROM students
INNER JOIN courses ON students.student_id = courses.student_id
''')
joined_data = cursor.fetchall()

print("Students and their courses (INNER JOIN):")
for name, course in joined_data:
    print(f"{name} is enrolled in {course}")
```

## LEFT JOIN Example

```python
# LEFT JOIN to show all students and their courses, even if they have none
cursor.execute('''
SELECT students.name, courses.course_name
FROM students
LEFT JOIN courses ON students.student_id = courses.student_id
''')
left_join_data = cursor.fetchall()

print("All students with their courses (including those without courses):")
for name, course in left_join_data:
    print(f"{name}: {course if course else 'No course enrolled'}")
```

## Creating and Using Views

```python
# Creating a view to simplify repeated queries
cursor.execute('''
CREATE VIEW student_courses AS
SELECT students.name, courses.course_name
FROM students
INNER JOIN courses ON students.student_id = courses.student_id
''')

# Querying the view
cursor.execute('SELECT * FROM student_courses')
view_data = cursor.fetchall()

print("View of students and their courses:")
for entry in view_data:
    print(entry)
```

## Nested Queries (Subqueries)
- **Nested queries** allow you to filter results based on the output of another query.
- In this example:
  - The inner query (`SELECT student_id FROM courses WHERE course_name = 'Mathematics'`) finds the `student_id`s of students enrolled in 'Mathematics'.
  - The outer query (`SELECT name FROM students WHERE student_id IN (...)`) fetches the names of those students.


```python
# Nested query to find the names of students who are enrolled in 'Mathematics'
cursor.execute('''
SELECT name
FROM students
WHERE student_id IN (
    SELECT student_id
    FROM courses
    WHERE course_name = 'Mathematics'
)
''')
simple_nested_query_result = cursor.fetchall()

print("Students enrolled in 'Mathematics':")
for student in simple_nested_query_result:
    print(student[0])
```
