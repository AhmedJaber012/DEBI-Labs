**1. How to "Pull" (Download) a Linux Image**



```python
docker pull ubuntu:latest
```

2. How to "Run" (Start) the Image

Once the image is downloaded, you create a "Container" (a running instance) from it:

```python
docker run -it --name student_lab ubuntu /bin/bash
```

-it: This makes it interactive so you can type commands.

--name student_lab: Gives your "Linux machine" a name.

ubuntu: The image you just downloaded.

/bin/bash: The shell (command line) you want to use.

to run again -> docker start -ai student_lab

**File & Directory Operations**

Navigate, create, and manage files and directories effectively.

**File Content & Manipulation**

View File Contents: `cat`, `head`, `tail`

```python
#create file
touch hello.csv

echo "1,Ahmed,A" >> hello.csv

echo "2,Sara,B" >> hello.csv

# Show first 10 lines
head filename

# Show last 10 lines
tail filename

# all file
cat filename

# find file
find / -name "hello.csv"
```

**Copy & Move: `cp`, `mv`**

Copy and move files and directories.

```python
# Copy file
cp source.txt destination.txt

# Copy directory recursively
cp -r sourcedir/ destdir/

# Move/rename file
mv oldname.txt newname.txt

# Move to different directory

mv file.txt /path/to/destination/
```

**File Permissions: `chmod`, `chown`**
Modify file permissions and ownership.

```python
# Change permissions (numeric)
chmod 755 filename
# Add execute permission
chmod +x script.sh
# Change ownership
chown user:group filename

# Change ownership recursively
chown -R user:group directory/
```

2. What the Letters Mean

- r (Read): You can open the file to see the data (e.g., using cat or less).

- w (Write): You can change the data, rename the file, or delete it.

- x (Execute): You can run the file as a program or script.

3. The Math (Numeric Mode)

In Linux administration (and within Docker), we often use numbers to change these permissions. Each letter has a value:

r = 4

w = 2

x = 1

Your file rw- r-- r-- is 644:

Owner: 4(r)+2(w)=6

Group: 4(r)+0=4

Others: 4(r)+0=4
