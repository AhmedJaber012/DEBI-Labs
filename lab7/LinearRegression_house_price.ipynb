```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

# ==============================
# 1️⃣ Load Data
# ==============================
df = pd.read_csv("/content/house_prices.csv")

print("Dataset preview:")
print(df.head(), "\n")
```

```python

# ==============================
# 2️⃣ Features & Target
# ==============================
X = df.drop("price", axis=1) # df.iloc[:,:5]
y = df["price"] # df.iloc[:,-1]
```

```python
X.shape # featues
```

```python
y.shape # target
```

```python
# ==============================
# 3️⃣ Train / Test Split
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

```python

# ==============================
# 4️⃣ Train Model
# ==============================
model = LinearRegression()
model.fit(X_train, y_train)
```

```python


# ==============================
# 5️⃣ Model Parameters
# ==============================
print("Model coefficients:")
for feature, coef in zip(X.columns, model.coef_):
    print(f"{feature}: {coef:.2f}")

print("Bias (Intercept):", model.intercept_, "\n")
```

```python

# ==============================
# 6️⃣ Predictions
# ==============================
y_pred = model.predict(X_test)
```

```python

# ==============================
# 7️⃣ Evaluation
# ==============================
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Evaluation Metrics:")
print("MSE :", mse)
print("RMSE:", rmse)
print("R2  :", r2, "\n")
```

```python
# ==============================
# 8️⃣ Test with New House
# ==============================
new_house = pd.DataFrame([{
    "size": 190,
    "bedrooms": 3,
    "bathrooms": 2,
    "age": 4,
    "location_score": 8
}])

predicted_price = model.predict(new_house)
print("Predicted price for new house:", int(predicted_price[0]))
```
